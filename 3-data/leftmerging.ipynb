```python
import pandas as pd
data_a = {'CustomerID': [1, 2, 3, 4, 5],
          'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
          'Region': ['North', 'South', 'North', 'West', 'East']}
df_customers = pd.DataFrame(data_a)
df_customers 
```

```python
data_b = {'OrderID': ['X100', 'X101', 'X102', 'X103', 'X104', 'X105'],
          'Customer_ID': [1, 3, 3, 6, 2, 2],
          'Amount': [50.0, 75.0, 10.0, 100.0, 25.0, 90.0]}
df_orders = pd.DataFrame(data_b)
df_orders
```

```python
result_left = pd.merge(df_customers, df_orders, left_on='CustomerID', right_on='Customer_ID', how='left')
result_left


```

A left join in pandas keeps all rows from the left DataFrame and brings in matching rows from the right DataFrame. 

If there is no match, the result will contain NaN (missing values).

All customers are kept (from df_customers — the left table).
Matching orders from df_orders are added.
If a customer has:

- multiple orders → multiple rows appear (e.g., Bob and Charlie)
- no orders → NaN is shown (e.g., David, Eve)
